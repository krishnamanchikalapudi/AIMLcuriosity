## Download dataset 
Download UC Irvine's concrete slump dataset to local folder '/data'

```python
import os
import urllib.request

# BEGIN: fix Python or Notebook SSL CERTIFICATE_VERIFY_FAILED
import ssl
if (not os.environ.get('PYTHONHTTPSVERIFY', '') and getattr(ssl, '_create_unverified_context', None)):
    ssl._create_default_https_context = ssl._create_unverified_context
# END: fix Python or Notebook SSL CERTIFICATE_VERIFY_FAILED

def retrieve_data(remote_file, local_file):
    if not os.path.exists(local_file) and not os.path.isfile(local_file):
        urllib.request.urlretrieve(remote_file, local_file)
        print(f"File: {local_file} download complete")
    else:
        print(f"File: '{local_file}' already exists!")
```

```python
remote_filename_concrete = 'https://archive.ics.uci.edu/ml/machine-learning-databases/concrete/slump/slump_test.data'
local_filename_concrete = './data/concrete_slump.csv'
retrieve_data(remote_filename_concrete, local_filename_concrete)
```

## Print the Fashion dataset

X_train_full has 60000 fashion items. So, printing all records in rows: 4000 and cols: 15

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

import time

n_rows = 15 # 4000
n_cols = 15
plt.figure(figsize=(n_cols * 1.2, n_rows * 1.2))
startTime = time.perf_counter() 
for row in range(n_rows):
    # print(f"row: {row} processing")
    for col in range(n_cols):
        index = n_cols * row + col
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(X_train_full[index], cmap="binary", interpolation="nearest")
        plt.axis('off')
        plt.title(class_names[y_train_full[index]], fontsize=12)

endTime = time.perf_counter() 
totalTime = (endTime - startTime)
print (f"Total records ({n_rows * n_cols}) processing time: {format(totalTime, '6.3f')} seconds or {format(totalTime / 60, '6.3f')} minutes\n\n")

plt.subplots_adjust(wspace=0.2, hspace=0.5)
plt.show()
```

# 5. Create Dataset

```python
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

def print_dataset(ds, quantity=5):
    for el in ds.take(quantity):
        print(el)


data = list(range(100))
dataset = tf.data.Dataset.from_tensor_slices(data)

# printing dataset
print_dataset(dataset)
```

## Shuffle dataset

```python

shuffled_dataset = dataset.shuffle(10)
print_dataset(shuffled_dataset)
```

```python
# Shuffle dataset
shuffled_dataset = dataset.shuffle(20)
print_dataset(shuffled_dataset)
```

## Batch dataset

```python
batch_dataset = dataset.batch(10)
print_dataset(batch_dataset)
```

```python
batch_dataset = dataset.batch(12, drop_remainder=True)
print_dataset(batch_dataset)
```

```python
batch_shuffled_dataset = dataset.batch(10).shuffle(10)
print_dataset(batch_shuffled_dataset)
```

```python
print_dataset(batch_shuffled_dataset)
```

```python
shuffled_batch_dataset = dataset.shuffle(10).batch(10)
print_dataset(shuffled_batch_dataset)
```

```python
print_dataset(shuffled_batch_dataset)
```

## Cache batch

```python
cached_batch = dataset.batch(10).cache()
print_dataset(cached_batch)
```

```python
print_dataset(cached_batch)
```

## Cache Shuffle

```python
cached_shuffle = dataset.shuffle(10).cache()
print_dataset(cached_shuffle)
```

```python
print_dataset(cached_shuffle)
```
